# Studying the Impact of COVID-19 on Australian Real Estate Investment Trusts

```python
import pandas as pd
SCG = pd.read_csv("SCG.csv")
SCP = pd.read_csv("SCP.csv")
VCX = pd.read_csv("VCX.csv")
ASX = pd.read_csv("ASX.csv")
ALU = pd.read_csv("ALU.csv")
TNE = pd.read_csv("TNE.csv")
XRO = pd.read_csv("XRO.csv")

stocks = [SCG, SCP, VCX, ALU, TNE, XRO]
```

```python
for stock in stocks:
    stock['Return'] = (stock['Close'] - stock['Close'].shift(1))/stock['Close'].shift(1)
    # stock['Date'] = pd.to_datetime(stock['Return'], infer_datetime_format = True)
ASX['Return'] = (ASX[' Close'] - ASX[' Close'].shift(1))/ASX[' Close'].shift(1)
```
